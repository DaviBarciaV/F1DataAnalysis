```python
tmp_df = df.query('`Pole positions` > 0')[['Piloto', 'Pole positions', 'Corridas ganhas', 'Campeão']]

color_discrete_map={False: dark_gray_color, True: highlight_color}

fig = px.scatter(
    tmp_df,
    x='Pole positions',
    y='Corridas ganhas',
    color='Campeão',
    trendline='ols',
    color_discrete_map=color_discrete_map
)

fig.update_layout(
    title='<b>Correlação entre pole position e vitórias</b>',
    yaxis_title='Vitórias #',
    xaxis_title="Pole Position #"
)

for index, row in tmp_df.query('`Pole positions` >= 50').iterrows():
    driver_name = row['Piloto'].split(' ')[-1]
    fig.add_annotation(x=row['Pole positions'], y=row['Corridas ganhas'], text=driver_name, showarrow=False, yshift=15)

fig.show()
```

```python
drivers = ['Lewis Hamilton', 'Michael Schumacher', 'Sebastian Vettel', 'Ayrton Senna']
drivers_filter = df['Piloto'].isin(drivers)
interest_cols = ['Taxa de Pole', 'Taxa de Vitória', 'Taxa de Pódio', 'Taxa de Volta Rápida']
champions_df = df[drivers_filter][['Piloto'] + interest_cols].reset_index(drop=True)
champions_df['Piloto'] = champions_df['Piloto'].apply(lambda x: x.split(' ')[1])

fig = px.parallel_coordinates(
    champions_df,
    dimensions=interest_cols,
    color=champions_df.index,
    range_color=[0.0, 3.0],
    color_continuous_scale=[
        (0.00, '#ff595e'), (0.25, '#ff595e'),
        (0.25, '#ffca3a'), (0.5, '#ffca3a'),
        (0.5, '#8ac926'),  (0.75, '#8ac926'),
        (0.75, '#1982c4'),  (1.00, '#1982c4')
    ],
    title='Hamilton, Schumacher, Vettel e Senna'
)

fig.update_layout(
    coloraxis_colorbar=dict(
        title=None,
        tickvals=[0, 1, 2, 3],
        ticktext=champions_df['Piloto'],
        lenmode="pixels",
        len=150,
    )
)

fig.show()
```
